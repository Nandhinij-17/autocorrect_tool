```python
from textblob import TextBlob
from transformers import pipeline
```

```python
def textblob_correct(text):
    corrected = str(TextBlob(text).correct())
    return corrected
```

```python
print(textblob_correct("I havv a dreem"))
```

```python
print("Loading BERT Model...")

bert = pipeline(
    "fill-mask",
    model="bert-base-uncased"
)

print("BERT Loaded Successfully!")
```

```python
def bert_suggestions(sentence):

    results = bert(sentence)

    print("\nTop Suggestions:\n")

    for i, r in enumerate(results[:5], start=1):
        print(f"{i}. {r['token_str']} ({r['score']:.4f})")
```

```python
bert_suggestions("I drink a cup of [MASK] every morning.")
```

```python
def auto_correct(text):

    corrected = str(TextBlob(text).correct())

    print("\n" + "="*50)
    print("          AUTOCORRECT RESULT")
    print("="*50)

    print("\nOriginal Text:")
    print(text)

    print("\nTextBlob Correction:")
    print(corrected)

    print("\nBERT Suggestions:")

    try:

        results = bert(corrected)

        for i, r in enumerate(results[:5], start=1):
            print(f"{i}. {r['token_str']}")

    except:

        print("No [MASK] token found.")

    print("\nFinal Output:")
    print(corrected)

    print("="*50)
```

```python
auto_correct("I havv a dreem")
```

```python
def about_project():

    print("\n" + "="*50)
    print("             ABOUT PROJECT")
    print("="*50)

    print("Project Name:")
    print("AI Autocorrect Tool")

    print("\nTechnologies:")
    print("- Python")
    print("- TextBlob")
    print("- BERT")

    print("\nFeatures:")
    print("- Spell Correction")
    print("- Context Prediction")
    print("- User Friendly Menu")

    print("="*50)
```

```python
while True:

    print("\n" + "="*50)
    print("         AI AUTOCORRECT TOOL")
    print("="*50)

    print("1. Auto Correct Text")
    print("2. About Project")
    print("3. Exit")

    choice = input("\nEnter Choice: ")

    if choice == "1":

        text = input("\nEnter Sentence: ")
        auto_correct(text)

    elif choice == "2":

        about_project()

    elif choice == "3":

        print("\nThank You!")
        break

    else:

        print("\nInvalid Choice")
```
